```python
%pylab inline
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import scipy.fftpack

pylab.rcParams['figure.figsize'] = (15, 10)
```

```python
# Number of samplepoints
N = 600
dt = 1.0 / 800.0

A_dc = 10
A_0 = 1.5
A_1 = 0.5

freq_0 = 50.0
freq_1 = 80.0
```

```python
x = np.linspace(0.0, N*dt, N)
y = A_dc + A_0*np.sin(2.0*np.pi*freq_0*x) + A_1*np.sin(2.0*np.pi*freq_1*x)

_ = plt.plot(x, y)
plt.title('y(t)')
plt.xlabel('t')
plt.xlabel('y')
```

```python
Fk = np.linspace(0.0, 1.0/(2.0*dt), N)
Fy = scipy.fftpack.fft(y)

ax_1 = plt.subplot(2, 1, 1)
ax_1.tick_params(axis='y', colors='blue')
ax_1.plot(Fk[0:int(N/2)], 2.0/N*np.abs(Fy[0:int(N/2)]))
# plt.plot(Fk, 2.0/N*np.abs(Fy))

ax_2 = ax_1.twinx()
ax_2.tick_params(axis='y', colors='red')
ax_2.plot(Fk[1:int(N/2)], 2.0/N*np.abs(Fy)[1:int(N/2)], color='r')

Fk = Fk - (1/(4*dt))
Fy = scipy.fftpack.fftshift(Fy)

ax_3 = plt.subplot(2, 1, 2)
ax_3.tick_params(axis='y', colors='blue')
ax_3.plot(Fk, 2.0/N*np.abs(Fy))

# masking the DC part
Fy[int(N/2)] = 0
ax_4 = ax_3.twinx()

ax_4.tick_params(axis='y', colors='red')

# for child in ax_4.get_children():
#     if isinstance(child, matplotlib.spines.Spine):
#         child.set_color('#dddddd')

ax_4.plot(Fk[1:], 2.0/N*np.abs(Fy)[1:], color='r')
```
